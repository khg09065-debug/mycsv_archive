```python
## 특정 컬럼 지우기

import pandas as pd

# 1. CSV 파일 불러오기
df = pd.read_csv('total_coffee.csv')

# 2. 특정 컬럼 삭제 (예: '나이'와 '주소' 컬럼 삭제)
# columns 리스트에 삭제하고 싶은 컬럼명을 넣습니다.
df_deleted = df.drop(columns=['id'])

# 3. 결과 저장하기
df_deleted.to_csv('total_coffee.csv', index=False, encoding='utf-8-sig')

print("컬럼 삭제 완료!")
```

```python
## 특정 컬럼명 변경

import pandas as pd

# 1. CSV 파일 불러오기
df = pd.read_csv('coffee_투썸플레이스.csv')

# 2. 특정 컬럼명 변경 (예: 'name' -> '이름', 'age' -> '나이')
df = df.rename(columns={'매점명' : '매장명'})

# 3. 결과 저장
df.to_csv('coffee_투썸플레이스.csv', index=False, encoding='utf-8-sig')

print("컬럼 이름 변경 완료!")
```

```python
## 컬럼 값 변경

import pandas as pd

# 1. CSV 파일 불러오기
df = pd.read_csv('coffee_빽다방.csv')

# 2. '구' 컬럼 값을 '주소'의 첫 어절 앞 두 글자로 변경
# .str.split().str[0] : 공백으로 나눈 후 첫 번째 단어 선택
# .str[:2] : 그 단어의 앞 두 글자만 추출
df['구'] = df['주소'].str.split().str[0].str[:2]

# 3. 결과 확인 (상위 5개)
print(df[['매장명', '주소', '구']].head())

# 4. 수정된 파일 저장
df.to_csv('coffee_빽다방.csv', index=False, encoding='utf-8-sig')
```

```python
## 컬럼 추가 및 값 삽입

import pandas as pd

# 1. 파일 불러오기
df = pd.read_csv('coffee_폴바셋.csv')

# 2. '지역' 컬럼에 들어갈 값 추출 (주소의 첫 어절 앞 두 글자)
# 예: '부산 강서구...' -> '부산', '경기 남양주시...' -> '경기'
region_values = df['주소'].str.split().str[0].str[:2]

# 3. 맨 앞(0번 인덱스)에 '지역'이라는 이름으로 컬럼 추가
df.insert(0, '지역', region_values)

# 4. 결과 확인 (상위 5개 행)
print(df.head())

# 5. 수정된 내용을 새 파일로 저장
df.to_csv('coffee_폴바셋.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

df = pd.read_csv('coffee_이디야.csv')

# 현재 컬럼 확인
print("변경 전:", df.columns)

# 1. 원하는 순서대로 컬럼명을 리스트에 나열합니다.
new_order = ['지역', '매장명', '주소' , '위도', '경도']
df = df[new_order]

# 2. 결과 확인
print("변경 후:", df.columns)

# 3. 저장
df.to_csv('coffee_이디야.csv', index=False, encoding='utf-8-sig')
```
