```python
import pandas as pd

# 1. 파일 불러오기
file_name = 'total_coffee.csv'
df = pd.read_csv(file_name)

# 2. '지역' 컬럼에 들어있는 고유한 값(Unique) 확인
unique_regions = df['지역'].unique()

# 3. 각 지역별 데이터 개수(Value Counts) 확인
region_counts = df['지역'].value_counts(dropna=False) # 결측치(NaN)도 포함해서 확인

print("======= '지역' 컬럼 상세 확인 =======")

# 전체 지역 리스트 출력
print(f"\n1. 발견된 지역 종류 ({len(unique_regions)}종):")
print(unique_regions)

# 지역별 데이터 개수 출력
print("\n2. 지역별 데이터 개수 (많은 순):")
print(region_counts)

# 결측치(빈 값) 확인
null_count = df['지역'].isnull().sum()
if null_count > 0:
    print(f"\n⚠️ 주의: '지역'이 비어있는 데이터가 {null_count}건 발견되었습니다.")
else:
    print("\n✅ 모든 데이터에 '지역' 정보가 입력되어 있습니다.")

print("====================================")
```

```python
import pandas as pd

# 1. 파일 불러오기
file_name = 'total_coffee.csv'
df = pd.read_csv(file_name)

# 2. 제거 전 데이터 개수 확인
before_count = len(df)

# 3. '지역' 컬럼이 NaN(결측치)인 행 제거
# subset=['지역']을 지정하여 오직 '지역' 컬럼의 결측치만 기준으로 삼습니다.
df = df.dropna(subset=['지역'])

# 4. 제거 후 데이터 개수 확인
after_count = len(df)
removed_count = before_count - after_count

# 5. 결과 저장 (기존 파일에 덮어쓰기)
# 인코딩은 한글 깨짐 방지를 위해 utf-8-sig를 사용합니다.
df.to_csv(file_name, index=False, encoding='utf-8-sig')

print(f"--- 결측치 제거 작업 완료 ---")
print(f"삭제 전 전체 행 수: {before_count}개")
print(f"삭제된 결측치 행 수: {removed_count}개")
print(f"최종 남은 행 수: {after_count}개")
print(f"결과가 '{file_name}'에 성공적으로 반영되었습니다.")
```

```python
import pandas as pd
import os

# 1. 파일 불러오기
input_file = 'total_coffee.csv'

try:
    df = pd.read_csv(input_file)
    
    # 2. 브랜드명 컬럼의 모든 고유값(Unique) 가져오기
    # 데이터에 결측치가 있을 경우를 대비해 문자열로 변환하고 처리합니다.
    brands = df['브랜드명'].dropna().unique()
    
    print(f"분석된 브랜드 수: {len(brands)}개")
    print("-" * 30)

    # 3. 브랜드별로 데이터를 필터링하여 저장
    for brand in brands:
        # 파일명에 사용할 수 없는 특수문자 제거 (안전장치)
        safe_brand_name = str(brand).replace('/', '_').replace('\\', '_').strip()
        output_file = f'coffee_{safe_brand_name}.csv'
        
        # 해당 브랜드의 데이터만 추출
        brand_df = df[df['브랜드명'] == brand]
        
        # 파일 저장 (utf-8-sig 인코딩으로 한글 깨짐 방지)
        brand_df.to_csv(output_file, index=False, encoding='utf-8-sig')
        
        print(f"저장 완료: {output_file} (총 {len(brand_df)}건)")

    print("-" * 30)
    print("모든 브랜드별 분할 작업이 성공적으로 완료되었습니다!")

except FileNotFoundError:
    print(f"오류: '{input_file}' 파일을 찾을 수 없습니다. 파일명을 확인해 주세요.")
except Exception as e:
    print(f"오류 발생: {e}")
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

# 1. CSV 파일 로드
file_path = 'total_coffee.csv'
df = pd.read_csv(file_path)

# 2. MySQL 연결 설정 (사용자 정보에 맞게 수정 필요)
user = 'root'      # MySQL 사용자명 (예: root)
password = '1234'  # MySQL 비밀번호
host = 'localhost'          # 호스트 (예: 127.0.0.1)
port = '3306'               # 포트 번호
db_name = 'coffee_store'    # 대상 데이터베이스

# 3. 데이터베이스 생성 (이미 존재하면 무시)
# 먼저 MySQL 서버 자체에 연결하여 DB가 없으면 생성합니다.
temp_engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}")
with temp_engine.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {db_name}"))
    print(f"데이터베이스 '{db_name}'가 준비되었습니다.")

# 4. SQLAlchemy 엔진 생성 (coffee_store DB 연결)
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{db_name}?charset=utf8mb4")

# 5. 데이터프레임을 MySQL 테이블로 저장
# - name: 생성할 테이블 이름 ('the_venti')
# - if_exists: 'fail' (이미 있으면 오류), 'replace' (기존 삭제 후 생성), 'append' (기존 테이블에 추가)
# - index: False (데이터프레임의 인덱스는 컬럼으로 넣지 않음)
try:
    df.to_sql(name='coffee_chain', con=engine, if_exists='replace', index=False)
    print(f"테이블 'coffee_chain'에 총 {len(df)}건의 데이터가 성공적으로 저장되었습니다.")
except Exception as e:
    print(f"데이터 저장 중 오류 발생: {e}")

# 연결 종료
engine.dispose()
```
