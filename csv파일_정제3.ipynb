```python
import pandas as pd
import os

# 1. 합칠 브랜드의 정확한 순서 리스트 정의
brand_order = [
    "스타벅스", "투썸플레이스", "이디야", "메가커피", "빽다방", 
    "컴포즈커피", "더벤티", "공차", "메머드커피", "디저트39", 
    "할리스", "폴바셋", "파스쿠찌", "하삼동커피", "던킨도너츠"
]

# 2. 개별 데이터프레임을 담을 리스트
all_dfs = []

# 3. 리스트 순서대로 파일 읽기
for brand in brand_order:
    file_name = f"coffee_{brand}.csv"
    
    # 파일이 실제로 존재하는지 확인
    if os.path.exists(file_name):
        print(f"[{file_name}] 파일을 불러오는 중...")
        df = pd.read_csv(file_name)
        
        # 데이터프레임 리스트에 추가
        all_dfs.append(df)
    else:
        # 파일이 없을 경우 알림 출력 (건너뜀)
        print(f"경고: [{file_name}] 파일이 폴더에 없습니다. 건너뜁니다.")

# 4. 모든 데이터프레임을 하나로 통합 (누적)
if all_dfs:
    # ignore_index=True를 주어 인덱스를 0부터 다시 새로 부여합니다.
    total_df = pd.concat(all_dfs, ignore_index=True)
    
    # 5. 최종 결과 저장
    output_name = "total_coffee.csv"
    total_df.to_csv(output_name, index=False, encoding='utf-8-sig')
    
    print("-" * 30)
    print(f"통합 완료! 총 {len(total_df)}개의 행이 합쳐졌습니다.")
    print(f"결과 파일: {output_name}")
else:
    print("합칠 수 있는 파일이 하나도 없습니다.")
```
